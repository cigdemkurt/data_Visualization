```python
import pandas as pd
import matplotlib.pyplot as plt

# Veri setini yükleme (örnek olarak csv dosyasından okuma)
data = pd.read_csv('C:\\Users\\Asus\\Desktop\\Supermarket.csv')  # Veri setinin adını ve yolunu dosyanıza göre güncelleyin

# Veri setinin ilk birkaç satırını görüntüleme (opsiyonel)
print(data.head())

# Örnek bir görselleştirme: Satış miktarlarının ürün bazında gösterilmesi
sales_by_product = data.groupby('Product Category')['Unit Price'].sum().sort_values(ascending=False)[:10]

plt.figure(figsize=(10, 6))
sales_by_product.plot(kind='bar', color='skyblue')
plt.title('Best-Selling Products')
plt.xlabel('Product Category')
plt.ylabel('Quantitaty ordered new')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('C:\\Users\\Asus\\Desktop\\Supermarket.csv')  # Veri setinin adını ve yolunu dosyanıza göre güncelleyin

# Ülkelere göre toplam satın alma miktarlarını hesaplama
country_sales = data.groupby('City')['Unit Price'].sum().sort_values(ascending=False)

# Pasta grafiği oluşturma
plt.figure(figsize=(10, 8))
country_sales.plot(kind='pie', autopct='%1.1f%%', startangle=90, cmap='tab20b')
plt.title('the amount of purchases by city')
plt.ylabel('total amount purchased')  # y etiketini kaldırma
plt.axis('equal')  # Dairesel şeklin korunmasını sağlar
plt.legend(loc='best', bbox_to_anchor=(1, 1))  # En iyi konumda ve dışarıda bir legent oluşturur
plt.show()
```
